```python
import numpy as np

def extract_logprobs_from_vllm_output(logprob_list):
    """
    Extracts all `logprob` values from a vLLM-style output list.
    Each entry in the list is either `None` or a {token_id: Logprob} dict.
    """
    logprobs = []
    for entry in logprob_list:
        if isinstance(entry, dict):
            for logprob_obj in entry.values():
                if hasattr(logprob_obj, "logprob"):
                    logprobs.append(logprob_obj.logprob)
    return logprobs

def compute_logprob_stats(logprobs):
    if not logprobs:
        return {"mean_logprob": float("-inf"), "total_logprob": float("-inf"), "num_tokens": 0}
    return {
        "mean_logprob": np.mean(logprobs),
        "total_logprob": np.sum(logprobs),
        "num_tokens": len(logprobs)
    }

def print_stats(name, stats):
    print(f"\n=== {name} ===")
    for key, val in stats.items():
        print(f"{key}: {val:.4f}")

def compare_logprobs(logprobs_cot_raw, logprobs_nocot_raw):
    logprobs_cot = extract_logprobs_from_vllm_output(logprobs_cot_raw)
    logprobs_nocot = extract_logprobs_from_vllm_output(logprobs_nocot_raw)

    stats_cot = compute_logprob_stats(logprobs_cot)
    stats_nocot = compute_logprob_stats(logprobs_nocot)

    print_stats("Chain-of-Thought", stats_cot)
    print_stats("No Chain-of-Thought", stats_nocot)

    print("\n=== Comparison ===")
    print(f"Δ mean logprob: {stats_cot['mean_logprob'] - stats_nocot['mean_logprob']:.4f}")
    print(f"Δ total logprob: {stats_cot['total_logprob'] - stats_nocot['total_logprob']:.4f}")
```
